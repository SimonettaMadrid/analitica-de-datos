```python
#Importacion de librerias 
import pandas as pd
import numpy as np 
import matplotlib as plt
import seaborn as sns
```

```python
data1 = pd.read_csv("inmuebles2 reto.csv", skiprows=1)
```

```python
data2 = data1.drop(['0','Referencia'],axis=1)
data2
```

```python
filtro1 = data2[(data2['Fecha Alta']=='2007-04-29 00:00:00')|(data2['Fecha Alta']=='2004-01-02 00:00:00')]
filtro1
```

```python
filtro2 = data2[(data2['Tipo']=='Parking') |
                (data2['Tipo']=='Local') |
                (data2['Tipo']=='Industrial') |
                (data2['Tipo']=='Piso')]
filtro2
```

```python
filtro3 = data2[(data2['Provincia']=='Girona')&(data2['Superficie']>100)]
filtro3
```

```python
filtro4 = data2[(data2['Precio Venta']>1000000)|
                (data2['Vendedor']=='Carmen')|
                (data2['Vendedor']=='Joaquin')]
filtro4
```

```python
filtro5 = data2[~(data2['Provincia']=='Barcelona')]
filtro5
```

```python
filtro6 = data2[~(data2['Operación']=='Alquiler')&~(data2['Tipo']=='Oficina')]
filtro6
```

```python
filtro1.to_csv('filtro1_2.csv')
filtro2.to_csv('filtro2_2.csv')
filtro3.to_csv('filtro3_2.csv')
filtro4.to_csv('filtro4_2.csv')
filtro5.to_csv('filtro5_2.csv')
filtro6.to_csv('filtro6_2.csv')
```
