```python
#importar numpy y matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definimos 2 arreglos de dimension 100
ys= 200 + np.random.randn(100)
x= np.array(range(0,100,1))
```

```python
#graficamos
plt.plot(x,ys,'*')
plt.fill_between(x,ys,195,facecolor='b',alpha=0.6)
#titulo a grafica
plt.title('Grafica de visualizacion')
plt.xlabel('Eje X', size=16,)
plt.ylabel('Eje Y',size=16,)
plt.show()
#
```
